```python
championships=soup.find_all('div',class_="matchCard")


for i in range(len(championships)):
    championships_name=championships[i].contents[1].find('h2').text.strip()

    f=open("matches.txt",'a',encoding='utf-8')
    f.write(championships_name)
    
    matches = championships[i].contents[3].find_all('div', class_='teamsData')

    for j in range(len(matches)):
       teamA = matches[j].find('div', {'class':'teamA'}).text.strip()
       teamB = matches[j].find('div', class_='teamB').text.strip()
       scoreA = matches[j].find_all('span', class_='score')[0].text.strip()
       scoreB = matches[j].find_all('span', class_='score')[1].text.strip()
       time = matches[j].find('span', class_='time').text.strip()
       f.write(f'{teamA} {scoreA} - {scoreB} {teamB} \n\t\t {time}')
       f.write('\n')

    f.write("======================================================")
    f.write('\n')
    f.close()   
           
      
```
